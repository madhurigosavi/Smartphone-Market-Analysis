#Visualizing the Distribution of Categorical Columns

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot_categorical_distribution(column):
    plt.figure(figsize=(8,4))
    sns.countplot(data=df, x=column, order=df[column].value_counts().index, hue=column, palette='Set2', legend=False)
    plt.title(f'Distribution of {column}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
plot_categorical_distribution('gender')
plot_categorical_distribution('age_group')
plot_categorical_distribution('education_level')
plot_categorical_distribution('monthly_income')
plot_categorical_distribution('employment_status')
plot_categorical_distribution('current_brand')
plot_categorical_distribution('current_price_range')
plot_categorical_distribution('purchase_source')
plot_categorical_distribution('residential_area')
```

#Multivariate Analysis:

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_multivariate_bar(data, x, hue, col=None, palette="Set2", height=5, aspect=1.2, rotation=45, title=""):
    """
    Plots a multivariate count plot using seaborn's catplot.
    
    Parameters:
    - data: DataFrame
    - x: str (categorical column for x-axis)
    - hue: str (categorical column to color bars)
    - col: str (optional, small multiples by this column)
    - palette: str (Seaborn color palette)
    - height, aspect: for figure sizing
    - rotation: angle for x-axis labels
    - title: str (overall plot title)
    """
    g = sns.catplot(
        data=data,
        kind="count",
        x=x,
        hue=hue,
        col=col,
        palette=palette,
        height=height,
        aspect=aspect
    )
    
    g.set_titles("{col_name}" if col else "")
    g.set_axis_labels(x.replace("_", " ").title(), "Count")
    g.set_xticklabels(rotation=rotation)
    
    if title:
        plt.subplots_adjust(top=0.85)
        g.fig.suptitle(title, fontsize=16)
    
    plt.tight_layout()
    plt.show()
```

```python
plot_multivariate_bar(
    data=df,
    x="current_price_range",
    hue="employment_status",
    col="monthly_income",
    title="Smartphone Price Range by Income & Employment Status"
)
```

```python
plot_multivariate_bar(
    data=df,
    x="current_price_range",
    hue="monthly_income",
    col="age_group",
    title="Smartphone Price Range by Age Group & Income"
)
```
