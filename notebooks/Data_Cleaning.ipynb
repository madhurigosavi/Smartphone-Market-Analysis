```python
def clean_model_name(value):
    if pd.isna(value):
        return None

    value = str(value).strip().lower()

    # Common typo corrections
    replacements = {
        'realmi': 'realme',
        'xiomi': 'xiaomi',
        'huwai': 'huawei',
        'inqoo': 'iqoo',
        'iphon': 'iphone',
        'not ': 'note ',
        'samsumg': 'samsung',
        'galaxy  ': 'galaxy ',
        'vivov': 'vivo v',
        'samsung  galaxy': 'samsung galaxy',
        'pgo': 'poco',
        'eluga 17': 'eluga i7'
    }

    for typo, correct in replacements.items():
        value = value.replace(typo, correct)

    # remove double spaces and extra whitespace
    value = ' '.join(value.split())

    return value
```

```python
def clean_model_name(value):
    if pd.isna(value):
        return None

    value = str(value).strip().lower()

    # Fix common brand/model typos
    corrections = {
        'redemi': 'redmi',
        'redme': 'redmi',
        'realmi': 'realme',
        'oppe': 'oppo',
        'vivio': 'vivo',
        'iphon': 'iphone',
        'i phone': 'iphone',
        'samsumg': 'samsung',
        'vvo': 'vivo',
        'techna': 'tecno',
        'techno': 'tecno'
    }

    for typo, fix in corrections.items():
        value = value.replace(typo, fix)

    # Remove multiple spaces
    value = ' '.join(value.split())

    # Remove invalid/suspicious entries
    if value in ['....', '53', '2020', 'kharde', '1+', '21']:
        return None

    return value
```

```python
reason_map = {
    '1': 'Camera quality',
    '2': 'Battery life',
    '3': 'Processor performance',
    '4': 'Display quality',
    '5': 'Storage & RAM',
    '6': 'Price & affordability',
    '7': 'Brand reputation',
    '8': '5G support',
    '9': 'Software & UI Experience',
    '10': 'After-Sales Service'
}
```

```python
def map_manual_reasons(code_str):
    if pd.isna(code_str):
        return None

    # Replace separators if any are incorrect (e.g., semicolon)
    code_str = code_str.replace(';', ',')
    
    codes = [c.strip() for c in code_str.split(',')]

    mapped = [reason_map.get(code, f"Unknown-{code}") for code in codes if code]
    return ', '.join(mapped)
```

```python
def clean_reasons(text):
    if pd.isna(text):
        return None

    text = text.lower()

    cleaned = []

    if 'camera' in text or 'कॅमेरा' in text:
        cleaned.append('Camera quality')
    if 'battery' in text or 'बॅटरी' in text:
        cleaned.append('Battery life')
    if 'processor' in text or 'प्रोसेसर' in text:
        cleaned.append('Processor performance')
    if 'display' in text or 'डिस्प्ले' in text:
        cleaned.append('Display quality')
    if 'storage' in text or 'ram' in text or 'स्टोरेज' in text:
        cleaned.append('Storage & RAM')
    if 'price' in text or 'affordability' in text or 'किंमत' in text:
        cleaned.append('Price & affordability')
    if 'brand' in text or 'ब्रँड' in text:
        cleaned.append('Brand reputation')
    if '5g' in text:
        cleaned.append('5G support')
    if 'software' in text or 'ui' in text:
        cleaned.append('Software & UI Experience')
    if 'after-sales' in text or 'service' in text or 'विक्रीनंतर' in text:
        cleaned.append('After-Sales Service')

    return ', '.join(cleaned)
```
